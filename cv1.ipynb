```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('arrow.png')
plt.imshow(img)
plt.title('original image')
plt.show()

```

```python
#image crop

cropped=img[0:100,0:200];
plt.imshow(cropped);
plt.show()
```

```python
#resized

resized=cv2.resize(img,(100,100))
plt.imshow(resized);
plt.show()
```

```python
#thresold

gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
__, thresh=cv2.threshold(gray,0,180,cv2.THRESH_BINARY)
plt.imshow(thresh)
plt.show()
```

```python
#contour
contours,_=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
image=img.copy()
contourimage=cv2.drawContours(image,contours,-1,(0,255,0),2)


plt.imshow(contourimage)

plt.show()
```

```python
detector=cv2.SimpleBlobDetector_create()
img2=cv2.imread('ferrari.jpg')
keypoints=detector.detect(img2)
blobimage=cv2.drawKeypoints(img2,keypoints,None,(0,250,255),cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
blobrgb = cv2.cvtColor(blobimage, cv2.COLOR_BGR2RGB)

plt.imshow(blobrgb)
plt.show()
```
